```python
categorical_variables = ['x24', 'x29', 'x30']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    d_test[variable].fillna("Missing", inplace=True)
    # Create array of dummies
    dummies = pd.get_dummies(d_test[variable], prefix=variable)
    # Update X to include dummies and drop the main variable
    d_test = pd.concat([d_test, dummies], axis=1)
    d_test.drop([variable], axis=1, inplace=True)
```
