```python
categorical_variables = ['x24', 'x29', 'x30']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    d_train[variable].fillna("Missing", inplace=True)
    # Create array of dummies
    dummies = pd.get_dummies(d_train[variable], prefix=variable)
    # Update X to include dummies and drop the main variable
    d_train = pd.concat([d_train, dummies], axis=1)
    d_train.drop([variable], axis=1, inplace=True)
```

As we are here to  predict whether the customer has cancelled the subscription, we shall try the generic was of using the approch of using  Random Forest Classiefier as we are predicting a class value (0/1). 

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, oob_score=True, n_jobs=-1, random_state=42)
```

Here, I have an oob_score of 0.920237500, which is comparatively low. Here, I am predicting a model using all the the variables. I could have deleted the x_24_ columns but as we are unaware of the variable names, its difficult to analyse whether this day in x_29 column:
* are the days where the delivery was late
* the food quality was bad
* the menu not suitable to their diet preferences 
 These factors could also have played an important role in deciding whether the customer wants to continue his subscritpion with Green Plate or no. 

** Here, I am going to use the Random Forest Regressor. The model can  give us the best accurate results using the
correct combination of the parameters which tune the model. The various Tuning Prameters used in Random forest are: **
* n_estimators : Number of tree
* max_features (0.5) : This feature allows us to take 50% of variable in an individual run.   
* min_samples_leaf (3):Here, technically I should use a higher min_samples_leaf. But, here 3 is working fine for my prediction model.
* n_jobs (-1) : Here, I have put n_jobs as -1 where there are no restrictions on how many processors the engine can use. Putting n_jobs=-1 shall also speed  up the run process for our model.   
* random_state: (42)
